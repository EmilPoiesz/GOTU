```python
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt

data = np.load('maj3dim40freeze2_transformer_1_1e-05_5_adam_0.npz')
lst = data.files
```

```python
df_loss = pd.DataFrame({
    lst[1]: data[lst[1]],
    lst[2]: data[lst[2]],
    lst[3]: data[lst[3]]
}).astype('float64')
df_loss.plot.line()
```

```python
coefficients = ['c', 'x1', 'x2', 'x3', 'x1x2', 'x1x3', 'x2x3', 'x1x2x3']
df_coeff = pd.DataFrame(data[lst[4]].squeeze(), columns=[coefficients[i] for i in range(data[lst[4]].squeeze().shape[1])]).astype('float64')
```

```python
ax = df_coeff.plot.line(title='Coefficients learned')
ax.axhline(0)

ax.set_xlabel('Epochs')
ax.set_ylabel('Coefficient value')
```
